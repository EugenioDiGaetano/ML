```python
import numpy as np
import time
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2").to(device)
```

## With KV caching

```python
start = time.time()
y = model.generate(**tokenizer("What is machine learning?", return_tensors="pt").to(device), use_cache=True, max_new_tokens=1000, do_sample=True)
t = time.time() - start
print(tokenizer.batch_decode(y)[0])
print("---")
print(f"Generation time: {round(t, 3)} seconds")
```

## Without KV caching

```python
start = time.time()
model.generate(**tokenizer("What is machine learning?", return_tensors="pt").to(device), use_cache=False, max_new_tokens=1000, do_sample=True)
t = time.time() - start
print(f"Generation time: {round(t, 3)} seconds")
```
